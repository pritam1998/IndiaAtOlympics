```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

plt.style.use('ggplot')

df = pd.read_csv(r'C:\Users\pritam\Desktop\Data\olympic data\athlete_events.csv')
df.head()
```

```python
df = df[df.Season == 'Summer']
df_india = df[df.NOC == 'IND']
```

```python
df_india.groupby(['Year'])['ID'].count().plot(marker='o', ylabel='No. of Athletes', figsize=(12,6))
plt.show()
```

```python
df_india.groupby(['Year','Sex']).count()['ID'].unstack().plot(kind='bar', ylabel='Male v/s Female Athletes', figsize=(12,6), width=0.7)
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(12,6))

ax.plot(df.groupby(['Year'])['Event'].nunique(), color='r', label='Total Events')
ax.plot(df_india.groupby(['Year'])['Event'].nunique(), color='b', label='India events participated')
plt.legend()
plt.show()
```

```python
df_india.groupby(['Year','Medal']).count()['ID'].unstack().reset_index().plot(x='Year', kind='barh',
                                                                           stacked=True, ylabel='Medals', figsize=(12,6))
```
